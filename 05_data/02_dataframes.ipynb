# Mas filtros a tener en cuenta

### 1️⃣ Filtro por múltiples condiciones (AND)

Este es el que ya viste y es probablemente el más común.
Ejemplo: flores con sépalo mayor a 5 y pétalo mayor a 3.

- df[(df["sepal_length"] > 5) & (df["petal_length"] > 3)]


Lo que suelen preguntar en entrevistas:
“Filtra las filas donde dos condiciones se cumplan al mismo tiempo.”
Variación típica:

- df[(df["age"] > 30) & (df["salary"] > 50000)]

### 2️⃣ Filtro por valores específicos (isin)

Esto es muy usado cuando quieres filtrar varias categorías.
Ejemplo con el famoso Iris Dataset:

- df[df["species"].isin(["setosa", "virginica"])]

Esto es equivalente a:

- (df["species"] == "setosa") | (df["species"] == "virginica")


pero mucho más limpio.
En entrevistas lo preguntan así:
“Filtra las filas donde la columna pertenezca a una lista de valores.”
Ejemplo real de negocio:

- df[df["country"].isin(["Colombia", "Peru", "Chile"])]
